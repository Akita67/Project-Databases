```python
from tkinter import*
import import_ipynb
import pizzasql as ps
```

```python
#menu window from which an order is placed
def menu_window(window=None, username=None):
    if window is not None:
        window.destroy()
        
    window = Tk()
    
    #pizza frame
    pizza_frm = Frame(master= window)
    pizza_frm.grid(row=0, column=0,rowspan=2)

    pizza_list = ps.get_menu_listing("pizza")
    create_menu_list(pizza_frm, pizza_list, "pizza")
    
    #drink frame
    drink_frm = Frame(master= window)
    drink_frm.grid(row=0, column=1,)

    drink_list = ps.get_menu_listing("drink")
    create_menu_list(drink_frm, drink_list, "drinks")

    #dessert frame
    dessert_frm = Frame(master= window)
    dessert_frm.grid(row=1, column=1,)

    dessert_list = ps.get_menu_listing("dessert")
    create_menu_list(dessert_frm, dessert_list,"dessert") 

    #discount frame
    discount_frm = Frame(master=window)
    discount_frm.grid(row=3, column=0)
    discount_lbl = Label(master=discount_frm, text="enter a discount code")
    discount_lbl.grid(row=0, column=0)
    discount_ent = Entry(master = discount_frm)
    discount_ent.grid(row=1,column=0)

    #checkout button
    checkout_btn = Button(text="Checkout", command=lambda: account_window(window))
    checkout_btn.grid(row=3, column=1)

    window.mainloop()
##########################################

def checkout(pizza_list, drink_list, deesert_list, idCustomer, use_coupon):
    #check for pizza
    if sum(price for _, price in pizza_list) == 0:
        pass #popup on no pizza
    
    #else procede with order
    else:
        order_items = []
        for tup in pizza_list + drink_list + dessert_list:
            if tup[1] > 0:
                order_items.append(tup)
        
        reward_coupon = None
        ps.execute_order(order_items, idCustomer, use_coupon, reward_coupon)

#this function creates a list of items from a menu category.
#it also returns a list of Spinbox widgets to read the user input
def create_menu_list(target_frm, source_list, menu_list_name):
    #title
    name_lbl = Label(master = target_frm, text = menu_list_name)
    name_lbl.grid(row = 0, column = 0, columnspan = 3)
    
    #heading
    pizza_lbl = Label(master = target_frm, text = "pizza options")
    pizza_lbl.grid(row = 1, column = 0)
    
    price_lbl = Label(master = target_frm, text = "unit price")
    price_lbl.grid(row = 1, column = 1)
    
    quantity_lbl = Label(master = target_frm, text = "quantity")
    quantity_lbl.grid(row = 1, column = 2)
    
    #spbx_list = []
    #for every listing
    for i, item in enumerate(source_list):
        
        name_btn = Button(master= target_frm, text= item[0], fg = "blue", relief=FLAT, 
            command= lambda i=i: pizza_popup(source_list[i][0])) # i=i to bypass python closure
        name_btn.grid(row = i+3, column = 0)
        
        price_lbl = Label(master= target_frm, text= item[1])
        price_lbl.grid(row = i+3, column = 1)
        
        quantity_spbx = Spinbox(master= target_frm, from_ = 0,  to = 20)
        quantity_spbx.grid(row = i+3, column = 2)
        
        #return quantity_spbx to able to uptodate resuls with quanitity_spbx.get()
        #spbx_list.append(quantity_spbx)
        
        #return spbx_list
    
####popup to show ingredientsof pizza
def pizza_popup(pizza):
    
    #ingrident list
    ingredient_list = ps.pizza_ingredient(pizza)#returns None if the ingredient is not a pizza
    
    if ingredient_list is None:
        return
    #vegetarian
    is_veg = ingredient_list.pop()
    
    popup = Toplevel()
    
    ingredient_lbl = Label(popup, text = ",\n".join(ingredient_list))
    ingredient_lbl.grid(row=0, column=0)
    
    if is_veg:
        is_veg = "This pizza is vegetarian"
    else:
         is_veg = "This pizza is NOT vegetarian"
            
    veg_lbl = Label(popup, text=is_veg)
    veg_lbl.grid(row=1, column=0)
    
    ok_btn = Button(popup, text="OK", command= lambda: popup.destroy())
    ok_btn.grid(row=2,column=0)
```

```python
def login_window(window=None):
    
    if window is not None:
        window.destroy()
        
    window = Tk()
    window.title('Personal data API')
    
    #user name input
    user_lbl = Label(text="username")
    user_lbl.grid(row=0, column=0)
    user_ent = Entry()
    user_ent.grid(row=0, column=1)

    #password input
    password_lbl = Label(text="password")
    password_lbl.grid(row=1,column=0)
    
    password_ent = Entry(show="*")
    password_ent.grid(row=1,column=1)

    # login button
    login_btn = Button(window, text='Login', command=lambda: menu_window(window))
    login_btn.grid(row=2, column=1)
    
    # Register button
    register_btn = Button(window, text='Register', command=lambda: print("Register"))
    register_btn.grid(row=2,column=0)
    
    window.mainloop()
########################################## 
```

```python
def account_window(window=None, username=None):
    if window is not None:
        window.destroy()
    
    window = Tk()
    window.title('Personal data Page')
    

    #Orders column and associated widgets
    orders_frm = Frame(window)
    orders_frm.grid(row=0,column=0)
    orders_lbl = Label(orders_frm, text="orders")
    orders_lbl.grid(row=0,column=0)
    
    #orders_display_frame
    order_disp_frm = Frame(orders_frm)
    order_disp_frm.grid(row=1,column=0)
    
    order_list = [("12-10-2022 20:00", 32.99),
                  ("11-10-2022 21:23", 43.09),
                  ("11-10-2022 23:12", 4.20),
                  ("13-04-2022 04:34", 6.9)]
    
    display_orders(order_disp_frm,order_list)

    # Menu button
    menu_btn = Button(orders_frm,text='MENU',command=lambda: menu_window(window))
    menu_btn.grid(row=2,column=0)
    
    # Cancel button
    cancel_btn = Button(orders_frm, text='CANCEL ORDER', command=lambda: print("cancel"))
    cancel_btn.grid(row=3,column=0)
    
    #View button
    view_btn = Button(orders_frm, text='ViEW ORDER', command=lambda: print("view"))
    view_btn.grid(row=4,column=0)
    
    #Coupons frame and assciated widgets
    coupon_frm = Frame(window)
    coupon_frm.grid(row=0, column=1)
    coupon_lbl = Label(coupon_frm, text="coupons")
    coupon_lbl.grid(row=0,column=0)

    #Personal details frame and associated widgets
    details_frm = Frame(window)
    details_frm.grid(row=0,column=1)
    
    address_lbl = Label(details_frm, text="Delivery Address")
    address_lbl.grid(row=0,column=0)
    
    #a frame with all address details
    address_frm = Frame(details_frm)
    address_frm.grid(row=1,column=0)
    address_form(address_frm)

    # Save button
    save_btn = Button(details_frm, text='SAVE',command=lambda: (print("save")))
    save_btn.grid(row=2,column=0)
    
    window.mainloop()
##########################################
```

```python
# will be changed to only select one at a time
def display_orders(target_frm,source_list):
     #heading
    time_lbl = Label(master = target_frm, text = "time")
    time_lbl.grid(row = 0, column = 0)
    
    price_lbl = Label(master = target_frm, text = " total price")
    price_lbl.grid(row = 0, column = 1)
    
    #for every listing
    for i, item in enumerate(source_list):
        name_chb = Checkbutton(master= target_frm, text=item[0])        
        price_lbl = Label(master= target_frm, text= item[1])
        
        name_chb.grid(row = i+1, column = 0)
        price_lbl.grid(row = i+1, column = 1)
```

```python
def address_form(target_frm):
    #make column 1 three time wider than 0
    target_frm.columnconfigure(0, weight=1)
    target_frm.columnconfigure(1, weight=3)

    ###information fields and field names
    #street number
    number_lbl = Label(target_frm, text="street number")
    number_lbl.grid(row=0,column=0)
    number_ent = Entry(target_frm)
    number_ent.grid(row=0,column=1)

    #street addition
    addition_lbl = Label(target_frm, text="street number addition")
    addition_lbl.grid(row=1,column=0)
    addition_ent = Entry(target_frm)
    addition_ent.grid(row=1,column=1)

    #street name
    name_lbl = Label(target_frm, text="street name")
    name_lbl.grid(row=2,column=0)
    name_ent = Entry(target_frm)
    name_ent.grid(row=2,column=1)

    #most code
    code_lbl = Label(target_frm, text="post code")
    code_lbl.grid(row=3,column=0)
    code_ent = Entry(target_frm)
    code_ent.grid(row=3,column=1)
```
